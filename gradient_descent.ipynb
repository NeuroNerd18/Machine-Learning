```python
from sklearn.datasets import make_regression
```

```python
import numpy as np
```

```python
X, y = make_regression(n_samples = 4 , n_features= 1, noise = 80, n_informative = 1, random_state = 80)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X, y)
```

```python
reg.intercept_
```

```python
#now we apply gradient descent considering slope is fixed and we need to find b
#we assume b = 0
```

```python
m= 79.73
b=0
lr= 0.1

epochs = 10

for i in range (epochs):
  loss_slope= -2 * np.sum(y-m*X.ravel()-b)
  b = b - (lr * loss_slope)
  y_pred = m*X +b


  plt.plot (X, y_pred)

plt.scatter(X, y)
```
